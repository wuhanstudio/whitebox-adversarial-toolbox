# Load image

```python
import cv2
import matplotlib.pyplot as plt
```

```python
dog = cv2.imread('dog.jpg')
dog = cv2.cvtColor(dog, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(dog)
```

# MobileNetV2 SSD Lite Inference

```python
import torch
from what.models.detection.ssd.mobilenet_v2_ssd_lite import MobileNetV2SSDLite
```

```python
# Initialize the model
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
model = MobileNetV2SSDLite(pretrained=True, is_test=True, device=device)
```

```python
# Run inference
_, boxes, labels, probs = model.predict(dog, 10, 0.4)
```

# Display output

```python
from what.models.detection.utils.box_utils import draw_bounding_boxes
```

```python
output = draw_bounding_boxes(dog, boxes, labels, model.class_names, probs)
```

```python
plt.imshow(output)
```
